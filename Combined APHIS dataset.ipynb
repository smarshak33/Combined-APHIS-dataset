```python
import pandas as pd
```

```python
## run the display code here to make numbers easier to read
pd.options.display.float_format='{:,.0f}'.format
```

```python
df_species = pd.read_csv("https://raw.githubusercontent.com/smarshak33/Combined-APHIS-dataset/refs/heads/main/inspections-species.csv")
df_species
```

```python
df_inspections=pd.read_csv("https://raw.githubusercontent.com/smarshak33/Combined-APHIS-dataset/refs/heads/main/inspections.csv")
df_inspections
```

```python
df_species.info()
```

```python
df_inspections.info()
```

```python
## merge
df= pd.merge(df_inspections, df_species, on="hash_id")
df
```

```python
df.info()
```

```python
df.sample()
```

```python
df.query("common == 'DOG ADULT'")
```

```python
df1=df.query("common == 'DOG ADULT' or common == 'DOG PUPPY'")
df1
```

```python
##How many unique certifications there are
df1["web_certType"].nunique()
```

```python
## What are the unique certifications?
df1["web_certType"].unique()
```
