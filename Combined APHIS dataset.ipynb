```python
import pandas as pd
```

```python
## run the display code here to make numbers easier to read
pd.options.display.float_format='{:,.0f}'.format
```

```python
df_species = pd.read_csv("https://raw.githubusercontent.com/smarshak33/Combined-APHIS-dataset/refs/heads/main/inspections-species.csv")
df_species
```

```python
df_inspections=pd.read_csv("https://raw.githubusercontent.com/smarshak33/Combined-APHIS-dataset/refs/heads/main/inspections.csv")
df_inspections
```

```python
df_species.info()
```

```python
df_inspections.info()
```

```python
## merge, combined dataset, first df
df= pd.merge(df_inspections, df_species, on="hash_id")
df
```

```python
df.info()
```

```python
df.sample()
```

```python
df1=df.query("common == 'DOG ADULT' or common == 'DOG PUPPY'")
df1
```

```python
##How many unique certifications there are
df1["web_certType"].nunique()
```

```python
## What are the unique certifications?
df1["web_certType"].unique()
```

```python
##download csv
df1.to_csv("Combined APHIS Dataset (all).csv", index=False)
```

```python
## without filtering for Class A breeders specifically, it appear there are 6,229 unique inspected licensees that breed, deal, test, or research on dogs
df1["web_legalName"].str.strip().str.upper().drop_duplicates()
```

# Class A specific df

```python
## Out of all the common names with dogs, how many inspection reports include class a breeders?
df1_classA = df1.query("web_certType == 'Class A - Breeder'")
df1_classA
```

```python
##download csv, will be updated.
df1_classA.to_csv("Class A Breeders.csv", index=False)
```

### web_legalName = name of inspected licensee. Appears to be tied to the customerNumber. It appears that when APHIS updates this value, it is updated for all rows with the same customerNumber.

```python
## Number of unique licensees appears to be 4,643 class A unique licensees
df1_classA["web_legalName"].str.strip().str.upper().drop_duplicates()
```
